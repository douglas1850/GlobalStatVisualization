```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('countries of the world.csv')
#Get rid of warning messages
import warnings
warnings.filterwarnings('ignore')

import plotly.plotly as py
import plotly.graph_objs as go 
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True)
```

```python
df.head(10)
```

```python
data = dict(
        type = 'choropleth',
        locations = df['Country'],
        locationmode = 'country names',
        z = df['GDP ($ per capita)'],
        text = df['Country'],
        colorbar = {'title' : 'GDP Per Capita in US $'},
      ) 
```

```python
layout = dict(
    title = 'Global GDP Per Capita',
    geo = dict(
        showframe = False,
        projection = {'type':'Stereographic'}
    )
)
```

```python
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap)
```

```python
data = dict(
        type = 'choropleth',
        locations = df['Country'],
        locationmode = 'country names',
        z = df['Population'],
        text = df['Country'],
        colorbar = {'title' : 'Population'},
      ) 
```

```python
layout = dict(
    title = 'Population',
    geo = dict(
        showframe = False,
        projection = {'type':'Conic equal area'}
    )
)
```

```python
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap)
```
